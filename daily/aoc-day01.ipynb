```python
def rotate(dial, r):
    zeros = 0
    direction = r[0]
    steps = int(r[1:])
    if direction == 'L':
        dial -= steps
    else:
        dial += steps
    # need to remain between 0-100
    while dial > 99:
        dial -= 100
        zeros += 1
    while dial < 0:
        dial += 100
        zeros += 1
    # did we land on zero?
    if dial == 0 and direction == 'L':
        zeros += 1
    return dial, zeros
```

```python
rotate(50, "L68")
```

```python
rotate(82, "L30")
```

```python
rotate(52, "R48")
```

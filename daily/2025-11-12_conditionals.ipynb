```python
# given one word, how would we find its frequency?
frequencies = {}
```

```python
for word in words:
    frequencies[word] = 0
```

```python
for word in words:
    frequencies[word] = frequencies[word] + 1
```

```python
# now to print the frequencies!
for key, value in frequencies.items():
    print(key, value)
```

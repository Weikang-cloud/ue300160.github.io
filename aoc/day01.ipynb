```python
def rotate_dial(dial, r):
    # L10 means -10
    # R560 means +560
    # !respect the boundary [0-99]
    direction = r[0]
    steps = int(r[1:])

    if direction == 'L':
        dial = dial - steps
    else:
        dial = dial + steps

    if dial >= 100:
        dial -= 100
    elif dial < 0:
        dial += 100
    
    return dial
```

```python
rotations = ['L68', 'L30', 'R48', 'L5', 'R60', 'L55', 'L1', 'L99', 'R14', 'L82']
dial = 50
print(dial)
for rotation in rotations:
    dial = rotate_dial(dial, rotation)
    print(f'- The dial is rotated {rotation} to point at {dial}.')
```
